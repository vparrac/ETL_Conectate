```python
import pandas as pd
import numpy as np
import os

directory='.'
i = 0
frames = []
for filename in os.listdir(directory):
    if not filename.endswith(".csv") or filename.startswith('profiling'):
        continue
    

    categories = {'Mean H':[], 'Mean O':[], 'Mean C':[], 'Mean A':[], 'Mean I':[], 'Mean E':[]}
    missing_val = {'Missing values H':[], 'Missing values O':[], 'Missing values C':[], 'Missing values A':[], 'Missing values I':[], 'Missing values E':[]}



    print(filename)
    df = pd.read_csv(filename,sep=';', header=3, encoding= 'unicode_escape')

    columns = list(df.columns)
    selected_columns = []
        
        
    for col in columns:
        name = str(col)
        if '_'in name:
            missing = df[name].isnull().sum()
            mean = df[name].mean(axis=0, skipna = True)
            if name.startswith('H'):
                categories['Mean H'].append(mean)
                missing_val['Missing values H'].append(missing)
            if name.startswith('O'):
                categories['Mean O'].append(mean)
                missing_val['Missing values O'].append(missing)
            if name.startswith('C'):
                categories['Mean C'].append(mean)
                missing_val['Missing values C'].append(missing)
            if name.startswith('A'):
                categories['Mean A'].append(mean)
                missing_val['Missing values A'].append(missing)
            if name.startswith('I'):
                categories['Mean I'].append(mean)
                missing_val['Missing values I'].append(missing)
            if name.startswith('E'):
                categories['Mean E'].append(mean)
                missing_val['Missing values E'].append(missing)
                    
           

    mean_values = {}
    nums = filename.split('-')
    row_id = nums[len(nums)-1].split('.')
    crn = row_id[0]
        
    mean_values['CRN'] = crn
    for k in categories.keys():
        mean_values[k] = np.mean(categories[k])
        
    for k in missing_val.keys():
        mean_values[k] = np.sum(missing_val[k])
            
            
    newDf = pd.DataFrame(mean_values,index=[0])
    frames.append(newDf)
    
result = pd.concat(frames, ignore_index=True)
finalFrame = pd.DataFrame(result)
print(finalFrame)

frame = finalFrame.fillna('null')

print(frame)
frame.to_csv('profiling.csv', index = False)
```
